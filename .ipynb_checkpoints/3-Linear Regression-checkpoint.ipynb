```r
lm_predictor <- function(predictors){
    variables <- paste('crim ~', paste(predictors, collapse = ' + '))
    model <- do.call("lm", list(as.formula(variables), data=as.name("Boston")))
    return(summary(model)$coefficients[predictors,'Pr(>|t|)'])
}
```

```r
summary(lm(crim ~ chas, data = Boston))$coefficients
```

```r
sapply(names(Boston)[-1], lm_predictor)
```

```r
ggpairs(Boston)
```

```r
lm25 <- lm(crim ~ ., data = Boston)
summary(lm25)
```

```r
lm_predictor_coeff <- function(predictors){
    variables <- paste('crim ~', paste(predictors, collapse = ' + '))
    model <- do.call("lm", list(as.formula(variables), data=as.name("Boston")))
    return(summary(model)$coefficients[predictors,'Estimate'])
}
```

```r
comparison <- cbind(sapply(names(Boston)[-1], lm_predictor_coeff), summary(lm25)$coefficients[-1, 'Estimate'])
```

```r
colnames(comparison) <- c('uni', 'multi')
comparison
```

```r
plot(comparison[, 'uni'] ~ comparison[, 'multi'])
```

```r
lm_predictor_cubed <- function(predictors){
    variables <- paste('crim ~', paste(predictors, collapse = ' + '))
    variables_cube <- paste0('crim ~ ', predictors, ' + I(', predictors, '^2)', ' + I(', predictors, '^3)')
    model <- do.call("lm", list(as.formula(variables), data=as.name("Boston")))
    model_cubed <- do.call("lm", list(as.formula(variables_cube), data=as.name("Boston")))
    
    return(anova(model, model_cubed))
}
```

```r
sapply(names(Boston)[-1], lm_predictor_cubed)
```
